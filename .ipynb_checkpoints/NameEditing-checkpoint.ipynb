```python
def connected_components(pairs):
    components = []
    for a, b in pairs:
        for component in components:
            if a in component:
                for i, other_component in enumerate(components):
                    if b in other_component and other_component != component: # a, and b are already in different components: merge
                        component.extend(other_component)
                        components[i:i+1] = []
                        break # we don't have to look for other components for b
                else: # b wasn't found in any other component
                    if b not in component:
                        component.append(b)
                break # we don't have to look for other components for a
            if b in component: # a wasn't in in the component 
                component.append(a)
                break # we don't have to look further
        else: # neither a nor b were found
            components.append([a, b])
    return components
```

```python
replacements = []
# Get rid of pak if the reminder is a name
for i,n in enumerate(names_l):
    s = re.search(r'\bpak(?P<word>\w+)\b',n)
    if s:
        ending = s.group("word")
        if ending in names_l:
            replacements.append(n)
            names_l[i] = ending
            
# Get rid of pa if the reminder is a name
for i,n in enumerate(names_l):
    s = re.search(r'\bpa(?P<word>\w+)\b',n)
    if s:
        ending = s.group("word")
        if ending in names_l:
            replacements.append(n)
            names_l[i] = ending

print(set(replacements))
print(len(set(names_l)), len(names_l))
```

```python
# Get rid of the dot ., comma and apostrophe
names_l = [re.sub(r"\.","",n) for n in names_l]
names_l = [re.sub(r",","",n) for n in names_l]
names_l = [re.sub(r"'","",n) for n in names_l]
names_l = [n.strip() for n in names_l]
names_l = [re.sub('\s+', ' ', n) for n in names_l]
print(len(set(names_l)), len(names_l))
```

```python
# Now we consider removing information in parentheses
# To do this we check if what is in parentheses is in the set of names, if it is, we transform this into an equivalent
# name, if not, we delete it
dictionary = {}
for n in set(names_l):
    s = re.search(r'(?P<word1>.*)\((?P<word2>.*)',n)
    if s:
        w1 = s.group('word1')
        w1 = w1.strip()
        w1 = re.sub('\s+', ' ', w1)
        w2 = s.group('word2')
        w2 = re.sub(r'\)','',s.group('word2'))
        w2 = w2.strip()
        w2 = re.sub('\s+', ' ', w2)
        if w1 in names_l and w1!="":
            if n in dictionary.keys():
                dictionary[n].append(w1)
            else:
                dictionary[n]=[w1]
        elif w2 in names_l and w2!="":
            if n in dictionary.keys():
                dictionary[n].append(w2)
            else:
                dictionary[n]=[w2,w1]
        else:
            dictionary[n]=[w1]
equivalences = {}
for i,n in enumerate(names_l):
    if n in dictionary.keys():
        if len(dictionary[n]) == 1:
            names_l[i] = dictionary[n][0]
        else:
            names_l[i] = dictionary[n][0]
            if not n in equivalences.keys():
                equivalences[n] = dictionary[n]
print(equivalences)
print(len(set(names_l)), len(names_l))
```

```python
# Construct more equivalences with the / character
dictionary = {}
for n in set(names_l):
    s = re.search(r'(?P<word1>.*)/(?P<word2>.*)',n)
    if s:
        w1 = s.group('word1')
        w1 = w1.strip()
        w1 = re.sub('\s+', ' ', w1)
        w2 = s.group('word2')
        w2 = w2.strip()
        w2 = re.sub('\s+', ' ', w2)
        if (w1 in names_l and w1!="") or (w2 in names_l and w2!=""):
            dictionary[n]=[w1,w2]
        else:
            dictionary[n]=[w1]
for i,n in enumerate(names_l):
    if n in dictionary.keys():
        if len(dictionary[n]) == 1:
            names_l[i] = dictionary[n][0]
        else:
            if dictionary[n][0] in names_l:
                names_l[i] = dictionary[n][0]
            else:
                names_l[i] = dictionary[n][1]
            if not n in equivalences.keys():
                equivalences[n] = dictionary[n]
print(equivalences)
print(len(set(names_l)), len(names_l))
```

```python
from scipy.sparse.csgraph import connected_components
nodes = []
for d in equivalences.keys():
    for s in equivalences[d]:
        nodes.append(s)
nodes = list(set(nodes))
n = len(nodes)
nodes_dict = dict(zip(nodes,list(range(len(nodes)))))
adjacency = np.zeros((n,n))
for d in equivalences.keys():
    for s1 in equivalences[d]:
        for s2 in equivalences[d]:
            adjacency[nodes_dict[s1],nodes_dict[s2]] = 1
components = connected_components(adjacency)

names_components = [[]]*components[0]
nodes_dict = dict(zip(list(range(len(nodes))),nodes))
for i in range(len(nodes)):
    names_components[components[1][i]]=names_components[components[1][i]]+[nodes_dict[i]]
names_components
```
