```python
%matplotlib inline
import datetime
from dateutil import parser
import math

from geopandas import GeoDataFrame
import pandas as pd
import numpy as np
from shapely.geometry import Point
```

```python
# Read middleman data
middle_df = pd.read_excel('/Users/sergiocamelo/Dropbox/Sergio-Joann/Trader Data/Overview_formatted.xlsx')
print("Number of middlemen with location: %d" % (len(middle_df)))
```

```python
# Calculate capacity and add it to dataframe
middle_df['capacity'] = (middle_df['# of trucks 10ton'] *10+
                         middle_df['# of trucks 2.5ton'] *2.5+
                         middle_df['# of trucks 1ton'] *1)

middle_df['trucks'] = [{10:middle_df['# of trucks 10ton'].iloc[i],2.5:middle_df['# of trucks 2.5ton'].iloc[i],1:middle_df['# of trucks 1ton'].iloc[i]} for i in range(len(middle_df))]

def aux(truck_count):
    truck_caps = {}
    j = 0
    for cap in truck_count:
        if pd.notnull(truck_count[cap]):
            for i in range(int(truck_count[cap])):
                truck_caps['_'.join(['t',str(j)])] = cap
                j += 1
    return truck_caps
                
middle_df['trucks'] =  middle_df['trucks'].apply(aux)
middle_df
```

```python
# Load Mills dataset
mills_df = pd.read_excel('/Users/sergiocamelo/Dropbox/Sergio-Joann/Mill-Locations.xlsx')
# Create mill dictionary
mills_dic = {}
for index, row in mills_df.iterrows():
    mills_dic[row['Code']] = [row['Latitude'], row['Longitude']]
mills_dic
```

```python
# Replace mills lat lon
def constr_mills(row):
    coord = {}
    mill_list = row['Mills to whom they sell']
    if pd.notnull(mill_list):
        mill_list = mill_list.split(',')
        mill_list = [m.strip() for m in mill_list]
        for m in mill_list:
            coord[m] = mills_dic[m]
    return coord
middle_df['mills_coord'] = middle_df.apply(constr_mills, axis=1)
```
