```python
def path_to_format( path):
        new_path = []
        for n in path:
            if (n<len(N)):
                new_path.append("n_"+str(n))
            else:
                new_path.append("h_"+str(n-len(N)))
        return (new_path)

def cost_path(path):
    cost_mat = lower_bound.reduced_cost_mat_complete(vrp.distance,u,v)
    cost = 0
    for i in range(len(path)-1):
        cost+=cost_mat[path[i]][path[i+1]]
    return(cost)
```
