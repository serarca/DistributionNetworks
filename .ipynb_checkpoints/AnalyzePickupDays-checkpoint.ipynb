```python
# We start by solving the TSP for any given day
daily_results = {}
for c in clusters:
    daily_results[c] = {}
    # Extract the location of the mills and the depot
    middle_df_c = middle_df[middle_df.cluster == c]
    M_p = np.array([middle_df_c['mills_coord'].values[0]['SKIP']])
    trucks = middle_df_c['trucks'].values[0]
    k = len(trucks)
    H = ['h_'+str(i) for i in range(k)]
    H_p = np.array(middle_df_c[['Lat','Lon']])
    H_p = np.tile(H_p,(k,1))
    capacities = dict(zip(H,trucks.values()))
    mapping_trucks = dict(zip(H,trucks.keys()))
    
    for d in range(days):
        
        df_cluster_day = df_clusters[(df_clusters['pickup'] == d) & (df_clusters['cluster'] == c)]

        # Construct optimization problem
        n = len(df_cluster_day)
        m = 1
        if n!=0:
            N = ['n_'+str(i) for i in range(n)]
            M = ['m_'+str(i) for i in range(m)]
            # This maps the depots H and nodes N to the truck_day and the farmers index, so we can
            # recover this information later
            mapping = {h:(mapping_trucks[h],d, c) for h in H}
            for i,ind in enumerate(df_cluster_day.index.values):
                mapping[N[i]] = ind
            quantities = {f: df_cluster_day['high'].values[i] for i,f in enumerate(N)}
            N_p = np.array(df_cluster_day[['lat','lon']])
            VRP_daily = VRPClass.VRP(H, N, H_p, N_p, quantities, capacities, 'geo', M, M_p)
            sol = VRP_daily.LinearSolver()
            daily_results[c][d] = {'value':VRP_daily.total_distance(sol), 
                                   'route':sol, 
                                  'mapping':mapping}
            
```

```python
# These are the exact solutions for the current allocations
added_distances = {c:np.sum([daily_results[c][i]['value'] for i in daily_results[c].keys()]) for c in clusters}
added_distances
```

```python
# Results temporal
results_temporal = {}

for c in clusters:
    print("cluster: %d" % (c))
    # Extract the location of the mills and the depot
    middle_df_c = middle_df[middle_df.cluster == c]
    M_p = np.array([middle_df_c['mills_coord'].values[0]['SKIP']])
    # Extract trucks from the data
    trucks = middle_df_c['trucks'].values[0]
    
    k = len(trucks) * days
    H = ['h_'+str(i) for i in range(k)]
    H_p = np.array(middle_df_c[['Lat','Lon']])
    H_p = np.tile(H_p,(k,1))
    
    # Construct the capacities
    capacities = dict(zip(H,trucks.values() * days))    
    
    df_cluster = df_clusters[(df_clusters['cluster'] == c)]
    
    
    # Construct optimization problem
    n = len(df_cluster)
    m = 1

    N = ['n_'+str(i) for i in range(n)]
    M = ['m_'+str(i) for i in range(m)]
    
    # Construct the mapping
    mapping = dict(zip(H,zip(trucks.keys() * days,np.array([[i]*len(trucks) for i in range(days)]).flatten())))
    mapping = {h:(mapping[h][0],mapping[h][1],c) for h in H}
    for i,ind in enumerate(df_cluster.index.values):
        mapping[N[i]] = ind
    
    quantities = {f: df_cluster['high'].values[i] for i,f in enumerate(N)}
    N_p = np.array(df_cluster[['lat','lon']])
    VRP_daily = VRPClass.VRP(H, N, H_p, N_p, quantities, capacities, 'geo', M, M_p)
    heuristic = VRP_daily.heuristic(VRP_daily.start_heuristic())
    print(VRP_daily.total_distance(heuristic))
    while True:
        cont = str(input("Continue?: "))
        if cont == 'y':
            sol = VRP_daily.LinearSolver(heuristic)
            heuristic = VRP_daily.heuristic(sol)
        else:
            break
    results_temporal[c] = {'value':VRP_daily.total_distance(heuristic),'route':heuristic, 'mapping':mapping}
```

```python
routes = {}
# For any given truck, day and cluster, we would like the distance traveled
# Construct the whole solution for each strategy
routes['current'] = {}
for c in clusters:
    for d in range(days):
        if d in daily_results[c].keys():
            sol = daily_results[c][d]
            mapping = sol['mapping']
            route = sol['route']
            for h in route.keys():
                path = [mapping[p] for p in route[h]['route']]
                routes['current'][mapping[h]] = {'path':path, 'distance':route[h]['route_dist']}
routes['temporal'] = {}
for c in clusters:
    sol = results_temporal[c]
    mapping = sol['mapping']
    route = sol['route']
    for h in route.keys():
        path = [mapping[p] for p in route[h]['route']]
        routes['temporal'][mapping[h]] = {'path':path, 'distance':route[h]['route_dist']}
routes['spatial'] = {}
for d in range(days):
    sol = results_spatial[d]
    mapping = sol['mapping']
    route = sol['route']
    for h in route.keys():
        path = [mapping[p] for p in route[h]['route']]
        routes['spatial'][mapping[h]] = {'path':path, 'distance':route[h]['route_dist']}
```

```python
# Calculate for each truck and cluster the total distance, total load and 
load_dict = dict(zip(df_clusters.index.values, df_clusters.high.values))
truck_data = {}
for opt in ['current','temporal','spatial']:
    truck_data[opt] = {}
    for c in clusters:
        middle_df_c = middle_df[middle_df.cluster == c]
        trucks = middle_df_c['trucks'].values[0]
        for t in trucks.keys():
            truck_data[opt][(c,t)] = {}
            truck_data[opt][(c,t)]['distance'] = np.sum([(routes[opt][(t,d,c)]['distance'] if ((t,d,c) in routes[opt].keys()) else 0) for d in range(days)])
            truck_data[opt][(c,t)]['max_distance'] = np.max([(routes[opt][(t,d,c)]['distance'] if ((t,d,c) in routes[opt].keys()) else 0) for d in range(days)])
            truck_data[opt][(c,t)]['days'] = np.sum([((routes[opt][(t,d,c)]['distance']>0) if ((t,d,c) in routes[opt].keys()) else 0) for d in range(days)])
            truck_data[opt][(c,t)]['avg_distance'] = truck_data[opt][(c,t)]['distance']/truck_data[opt][(c,t)]['days']
            truck_data[opt][(c,t)]['capacity'] = trucks[t]
            truck_data[opt][(c,t)]['load'] = np.sum([(np.sum([load_dict[ind] for ind in routes[opt][(t,d,c)]['path']]) if ((t,d,c) in routes[opt].keys()) else 0) for d in range(days)])
            truck_data[opt][(c,t)]['avg_load'] = truck_data[opt][(c,t)]['load']/truck_data[opt][(c,t)]['days']
            truck_data[opt][(c,t)]['avg_load_perc'] = truck_data[opt][(c,t)]['avg_load']/truck_data[opt][(c,t)]['capacity']
```

```python
# Construct table to present
d = {}
opt = 'current'
for k in truck_data[opt].keys():
    for ty in truck_data[opt][k].keys():
        d[(ty,k[0],k[1])] = [truck_data['current'][k]['capacity'],truck_data['current'][k][ty],truck_data['temporal'][k][ty],truck_data['spatial'][k][ty]]
mux = pd.MultiIndex.from_tuples(d.keys()).set_names(['Measure', 'Cluster','Truck'])
df = pd.DataFrame(list(d.values()), index=mux, columns =['capacity','Current','Temporal','Spatial'] )
df = df.sort_index()
df.to_csv('/Users/sergiocamelo/Dropbox/Sergio-Joann/Meetings/Apr17-2018/comparison.csv')
```

```python
%matplotlib inline
import numpy as np
from scipy.spatial import distance_matrix
from scipy.spatial.distance import pdist
import geopy.distance


V = H + N

# Generate a distance function
def get_entry(a,v):
    t=v
    if a == 'h':
        return H_p[int(v),:]
    if a == 'm':
        return M_p[int(v),:]
    if a == 'n':
        return N_p[int(v),:]

def v_distance(v1, v2):
    return geopy.distance.vincenty(get_entry(*str.split(v1,'_')),get_entry(*str.split(v2,'_'))).m

```

```python
# Create distance dictionary
distance_dict = {}
for v1 in V:
    distance_dict[v1] = {}
    for v2 in V:
        if (v1 in N) and (v2 in H):
            distance_dict[v1][v2] = v_distance(v1,'m_0') + v_distance('m_0',v2)
        else:
            distance_dict[v1][v2] = v_distance(v1,v2)
```

```python
capacities = {h:10.0 for h in H}
```

```python
# START HEURISTIC
# We start with a heuristic where we assign to each truck, the closest truck with availability
assignment_dict = {h:{'nodes':[],'cap':0} for h in H}
for n in N:
    dist_to_trucks = [distance_dict[n][h] for h in H]
    ranking = np.argsort(dist_to_trucks)
    q = quantities[n]
    assign = False
    for r in ranking:
        truck = H[r]
        truck_capacity = capacities[truck]
        current_load = assignment_dict[truck]['cap']
        if current_load + q <= truck_capacity:
            l = assignment_dict[truck]['nodes']
            l.append(n)
            assignment_dict[truck]['cap'] = current_load + q
            assignment_dict[truck]['nodes'] = l
            assign = True
            break
    if assign == False:
        print("Problem infeasible")
assignment_nodes = {n:h  for h in assignment_dict.keys() for n in assignment_dict[h]['nodes']}
```

```python
# Create the routes of the heuristic
iterations = 100
solution_tsp = {}
for h in H:
    if len(assignment_dict[h]['nodes'])>0:
        nodes = set(assignment_dict[h]['nodes'] + [h])
        best_dist = float('inf')
        best_route = list(nodes)
        for i in range(iterations):
            nn_solver = NearestNeighbourSolver(nodes, distance_dict)
            new_route = nn_solver.run()
            new_dist = nn_solver.get_total_distance()
            if new_dist < best_dist:
                best_dist = new_dist
                best_route = new_route
            ni_solver = NearestInsertionSolver(nodes, distance_dict)
            new_route = ni_solver.run()
            new_dist = ni_solver.get_total_distance()
            if new_dist < best_dist:
                best_dist = new_dist
                best_route = new_route
        solution_tsp[h] = {'route':best_route}
    else:
        solution_tsp[h] = {'route':[]}

# Edit solution
for h in H:
    l = solution_tsp[h]['route']
    if len(solution_tsp[h]['route'])>0:
        l.pop()
        ind = l.index(h)
        l = l[ind+1:len(l)]+l[0:ind]
        solution_tsp[h]['route'] = l
    solution_tsp[h]['load'] = np.sum([quantities[n] for n in l])
    solution_tsp[h]['max_load'] = capacities[h]
        
```

```python
# Create attributes vector (with taboos)
attributes = {}
for truck in H:
    for node in N:
        attributes[(truck,node)] = {'tabu':0,'rho':0}
        


# Calculate load violation
def calc_load(route):
    return np.sum([quantities[n] for n in route]) 
                   
def total_load_violation(sol):
    return np.sum([np.max([calc_load(sol[h]['route']) - capacities[h],0]) for h in H])

def distance_path(path):
    return np.sum([distance_dict[path[i]][path[i+1]] for i in range(len(path)-1)])

def total_distance(sol):
    return np.sum([distance_path([h]+sol[h]['route']+[h]) for h in H])

def rho(sol):
    return np.sum([ attributes[(h,n)]['rho'] for h in H for n in sol[h]['route']])
```

```python
import copy

alpha = 0.015
lamb = 1
theta = int(7.5*np.log10(len(N)))
delta = 0.5
# Calculate current cost
dist = total_distance(solution_tsp)
load = total_load_violation(solution_tsp)
f = dist + alpha*load
rho_sum = rho(solution_tsp)
```

```python
best_dist = dist
best_sol = copy.deepcopy(solution_tsp)
for it in range(10000):
    partial_data = {
        'cost' : float('inf'),
        'truck' : -1,
        'new_truck' : -1,
        'index' : -1,
        'new_index' : -1,
        'node' : -1
    }

    # Find two attributes to change
    for truck in H:
        for node in solution_tsp[truck]['route']:
            complete_route = [truck]+solution_tsp[truck]['route']+[truck]
            ind_node = complete_route.index(node)

            dec_dist = (distance_dict[complete_route[ind_node-1]][complete_route[ind_node+1]]
                        - distance_dict[complete_route[ind_node-1]][complete_route[ind_node]]
                        - distance_dict[complete_route[ind_node+1]][complete_route[ind_node]])
            overload = np.max([solution_tsp[truck]['load'] - solution_tsp[truck]['max_load'],0])
            dec_overload = - np.min([overload, quantities[node]])

            for new_truck in H:
                if truck != new_truck:
                    overload = np.max([solution_tsp[new_truck]['load'] - solution_tsp[new_truck]['max_load'],0])
                    new_overload = np.max([solution_tsp[new_truck]['load'] + quantities[node] - solution_tsp[new_truck]['max_load'],0])
                    inc_overload_truck = new_overload - overload
                    change_overload = inc_overload_truck + dec_overload
                    new_complete_route = [new_truck]+solution_tsp[new_truck]['route']+[new_truck]
                    for i in range(len(new_complete_route)-1):
                        inc = (distance_dict[new_complete_route[i]][node] 
                               + distance_dict[new_complete_route[i+1]][node]
                               -distance_dict[new_complete_route[i]][new_complete_route[i+1]])
                        change_dist = inc + dec_dist

                        # Calculate the new cost
                        new_distance = (dist + change_dist)
                        new_load = (load + change_overload)
                        new_f = new_distance + alpha*new_load
                        new_rho = rho_sum - attributes[(truck,node)]['rho'] + attributes[(new_truck,node)]['rho']
                        new_penalty = lamb*new_distance*np.sqrt(len(N)*len(H))*new_rho
                        cost = new_f+new_penalty

                        if (cost < partial_data['cost']) and (attributes[(new_truck,node)]['tabu']==0 or new_f<f) :
                            partial_data = {
                                'cost' : cost,
                                'truck' : truck,
                                'new_truck' : new_truck,
                                'index' : ind_node,
                                'new_index' : i,
                                'node' : node,
                                'dist_gain' : change_dist,
                                'load_gain' : change_overload
                            }

    # Do the update of the solution
    old_path = solution_tsp[partial_data['truck']]['route']
    old_path.pop(partial_data['index'] - 1)
    solution_tsp[partial_data['truck']]['route'] = old_path
    new_path = solution_tsp[partial_data['new_truck']]['route']
    new_path.insert(partial_data['new_index'],partial_data['node'])
    solution_tsp[partial_data['new_truck']]['route'] = new_path
    
    # Update loads
    solution_tsp[partial_data['new_truck']]['load'] += quantities[partial_data['node']]
    solution_tsp[partial_data['truck']]['load'] -= quantities[partial_data['node']]

    # Update attributes
    attributes[(partial_data['new_truck'],partial_data['node'])]['rho'] += 1
    for a in attributes.keys():
        attributes[a]['tabu'] = np.max([attributes[a]['tabu']-1,0])
    attributes[(partial_data['truck'],partial_data['node'])]['tabu'] = theta
    # Also update global vars
    dist = total_distance(solution_tsp)
    load = total_load_violation(solution_tsp)
    rho_sum = rho(solution_tsp)
    f = dist + alpha*load

    if load>0:
        alpha = alpha * (1+delta)
    else:
        alpha = alpha / (1+delta)
        if dist<best_dist:
            best_dist = dist
            best_sol = copy.deepcopy(solution_tsp)
            print(it, best_dist)
```
